```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_global_video_game_sales.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe().T
```

```python
df.isna().sum()
```

```python
sns.lineplot(data=df,x="Rank",y="Global_Sales")
```

Above Plot indicates that if the Rank is Higher(i.e. 1) than the Global Sales is also Higher

```python
plt.title(f"Year V/s Global Sales")
sns.lineplot(data=df,x=df["Year"].where(df["Year"]!=0),y=df["Global_Sales"])
plt.tight_layout()
plt.show()
```

It indicates that the Sales of the Games was Highest in the Year between 1988-89 and 2nd Highest in between 1984-85

```python
sns.histplot(df["Year"].where(df["Year"]!=0))
plt.title("Year V/s Count")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=90)
```

It indicates that Highest number of Games were Released which had more sales in the Year 2007-8

```python
grouped_genre = df.groupby("Genre")["Global_Sales"].sum()
```

```python
plt.bar(grouped_genre.index,grouped_genre.values)
plt.ylabel("Global Sales (in millions)")
plt.title("Genre V/s Global Sales ")
plt.xlabel("Genre")
plt.xticks(rotation=90)
```

It shows that the Action game has the Highest overall Global Sales around 1150 millions and Sports is the 2nd Highest Global Sales

```python
top10_platforms = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)
top10_platforms
```

```python
sns.barplot(x=top10_platforms.values,y=top10_platforms.index)
plt.xlabel("Global Sales (in millions)")
plt.title("Top 10 Platform By Global Sales")
```

It shows that the Games based on the PlayStaion Two Platform has the Highest Global Sales, followed by Wii

```python
top10_publisher = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.bar(top10_publisher.index,top10_publisher.values)
plt.xticks(rotation=90)
plt.title("Top 10 Publsher by Global Sales")
plt.xlabel("Publisher")
plt.ylabel("Global Sales(in millions)")
```

It inticates that Games Published by Nintendo has the Highest Sales followed by Electronic Arts and Sony Computer Entertainment

```python
df_var = df.describe().T.index
df_var = df_var.delete(0)
df_var = df_var.delete(0)
df_var = df_var.delete(4)
df_var
```

```python
plt.figure(figsize=(6,12))
i=1
for sales in df_var:
    plt.subplot(2,2,i)
    sns.scatterplot(x=df["Global_Sales"],y=df[sales])
    sns.regplot(data=df,x=df["Global_Sales"],y=df[sales])
    i+=1
plt.tight_layout()
plt.show()
```

It shows that NA, JP, EU and Other Sales have Correlation with Global Sales.

It can be seen that there are Some outliers in the above Plots

```python
i=1
plt.figure(figsize=(6,12))
for sales in df_var:
    plt.subplot(2,2,i)
    sns.boxplot(df[sales])
    plt.title(f"Rank V/s {sales}")
    i+=1

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(df['Global_Sales'])
plt.title("Rank V/s Global Sales")
plt.show()
```

```python
for sales in df_var:
    q1=np.percentile(df[sales],25)
    q3=np.percentile(df[sales],75)

    # print(f'Q1 : {q1}\nQ3 : {q3}')

    iqr=q3-q1
    # print(f'IQR : {iqr}')

    uw=q3+1.5*iqr
    lw=q1-1.5*iqr
    df.loc[df[sales]>uw, sales]=uw
    df.loc[df[sales]<lw, sales]=lw
    # print(f'Lower Whisker : {lw}\nUpper Whisker : {uw}')
i=1
plt.figure(figsize=(6,12))
for sales in df_var:
    plt.subplot(3,2,i)
    sns.boxplot(df[sales])
    plt.title(f"Rank V/s {sales}")
    i+=1

plt.tight_layout()
plt.show()
```

```python
q1=np.percentile(df["Global_Sales"],25)
q3=np.percentile(df["Global_Sales"],75)

iqr=q3-q1

uw=q3+1.5*iqr
lw=q1-1.5*iqr

df.loc[df["Global_Sales"]>uw, "Global_Sales"]=uw
df.loc[df["Global_Sales"]<lw, "Global_Sales"]=lw

sns.boxplot(df['Global_Sales'])
plt.title("Rank V/s Global Sales")
plt.show()
```

In above Box Plot Outliers have been Removed

```python
sns.heatmap(df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].corr(), annot=True, cmap='coolwarm')
```

The Above HeatMap gives the Correlation between different Sales of the Games
